VISUALIZATIONS

```python
plt.figure(figsize=(8,5))
school_data["State"].value_counts().plot(kind="bar")
plt.title("Number of Schools by State")
plt.xlabel("State")
plt.ylabel("Count")
plt.show()
```

```python

plt.figure(figsize=(6,6))
school_data["School_Type"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("School Type Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(school_data['Students_Count'], bins=15)
plt.title("Distribution of Students Count")
plt.xlabel("Number of Students")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(school_data['Teachers_Count'], bins=15)
plt.title("Distribution of Teachers Count")
plt.xlabel("Number of Teachers")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))

# Count of each School Type
school_counts = school_data['School_Type'].value_counts()

# Get colors from a Seaborn palette
colors = sns.color_palette("Set2", len(school_counts))  # vibrant colors

# Plot using Matplotlib
plt.bar(school_counts.index, school_counts.values, color=colors)

plt.title("Count of Schools by Type", fontsize=14)
plt.xlabel("School Type", fontsize=12)
plt.ylabel("Number of Schools", fontsize=12)
plt.show()
```

```python
# Get top 5 cities
top_cities = school_data['City'].value_counts().head(5)

plt.figure(figsize=(8,5))

# Use Seaborn palette for colors
colors = sns.color_palette("rainbow", len(top_cities))  # vibrant colors

# Plot bar chart with colors
top_cities.plot(kind='bar', color=colors)

plt.title("Top 5 Cities by Number of Schools", fontsize=14)
plt.xlabel("City", fontsize=12)
plt.ylabel("Number of Schools", fontsize=12)
plt.show()
```

```python
# Calculate average students per school type
avg_students = school_data.groupby('School_Type')['Students_Count'].mean()

plt.figure(figsize=(8,5))

# Use Seaborn palette
colors = sns.color_palette("Set2", len(avg_students))  # one color per school type

# Plot bar chart
avg_students.plot(kind='bar', color=colors)

plt.title("Average Students per School Type", fontsize=14)
plt.xlabel("School Type", fontsize=12)
plt.ylabel("Average Students", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(
    school_data[['Students_Count','Teachers_Count']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Between Students and Teachers")
plt.show()
```

INSIGHTS: 

- School Types: Dataset has both Govt and Private schools; Private schools are slightly more.

- Teacher Availability: Teacher count varies; some schools have fewer teachers despite high students.

- Student–Teacher Ratio: Not always balanced; some schools may need more teachers.

- City & State Distribution: Schools and student strength vary by location; urban areas tend to have larger schools.
